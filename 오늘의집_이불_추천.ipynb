## 색상유사도

```python
df['color_list_literal'] = df['color_list'].apply(lambda x : (' ').join(x))
df['color_list_literal']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# CountVectorizer로 학습시켜 428개 상품명에 대한 1541개 색상의 '색상 매트릭스' 생성
# 상품명 별 색상 인덱스 사전을 만들어준다
count_vect = CountVectorizer(min_df=0, ngram_range=(1, 2)) 
color_mat = count_vect.fit_transform(df['color_list_literal'])
print(color_mat.shape)
print(color_mat)
```

```python
# CountVectorizer로 학습시켜 428개 상품명에 대한 487개 색상의 '색상 매트릭스' 생성
count_vect2 = CountVectorizer(min_df=1, ngram_range=(1, 1))  # min_df: 단어장에 들어갈 최소빈도, ngram_range: 1 <= n <= 2
color_mat2 = count_vect2.fit_transform(df['color_list_literal'])
print(color_mat2.shape)
print(color_mat2)
```

```python
# 코사인 유사도에 의해 487개 이불 각각 유사한 이불들이 계산됨
from sklearn.metrics.pairwise import cosine_similarity
color_sim = cosine_similarity(color_mat, color_mat)

print(color_sim.shape)
print(color_sim[:10])
```

```python
# 자료를 정렬하는 것이 아니라 순서만 알고 싶다면 argsort
# 유사도가 높은 이불을 앞에서부터 순서대로 보여줌
color_sim_sorted_ind = color_sim.argsort()[:, ::-1]
print(color_sim_sorted_ind) 
```

```python
df['review_num'] = df['review_num'].str.replace(',','')
df['review_num'] = pd.to_numeric(df['review_num'])
df2['review_num'] = df2['review_num'].str.replace(',','')
df2['review_num'] = pd.to_numeric(df2['review_num'])
```

```python
# 상위 60%에 해당하는 리뷰수를 최소 리뷰수인 m으로 지정
C = df['rating'].mean()  # C :전체 상품에 대한 평균 평점
m = df['review_num'].quantile(0.6)
C = df2['rating'].mean() 
m = df2['review_num'].quantile(0.6)
```

```python
def weighted_vote_average(record): #가중평점
    v = record['review_num']  #리뷰수
    R = record['rating'] #별점
    
    return ( (v/(v+m)) * R ) + ( (m/(m+v)) * C )
```

```python
# 기존 데이터에 가중평점 칼럼 추가
df['weighted_rating'] = df.apply(weighted_vote_average, axis=1)
df2['weighted_rating'] = df2.apply(weighted_vote_average, axis=1)
```

```python
# 가중평점(weighted_vote) 기준으로 상품 순위 뽑아보기
df[['title','price','weighted_rating', 'rating', 'review_num', 'material', 'color_list_literal','Range']].sort_values('weighted_rating', ascending=False)
```
